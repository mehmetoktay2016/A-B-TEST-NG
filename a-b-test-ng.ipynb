CASE AB TEST



Analyze and Present A/B Test Results
Facebook recently introduced a new bidding type, “average bidding”, as an alternative to its exisiting bidding type, called “maximum bidding”. One of our clients, bombabomba.com, has decided to test this new feature and wants to conduct an A/B test to understand if average bidding brings more conversions than maximum bidding.
In this A/B test, bombabomba.com randomly splits its audience into two equally sized groups, e.g. the test and the control group. A Facebook ad campaign with “maximum bidding” is served to “control group” and another campaign with “average bidding” is served to the “test group”.
The A/B test has run for 1 month and bombabomba.com now expects you to analyze and present the results of this A/B test.

- Facebook Ad: An advertisement created by a business on Facebook that's served up to Facebook users.
- Impressions: The number of times an ad is displayed.
- Reach: The number of unique people who saw an ad.
- Website Clicks: The number of clicks on ad links directed to Advertiser’s website. Website Click Through Rate: Number of Website Clicks / Number of Impressions x 100 
- Cost per Action: Spend / Number of Actions
- Action: Can be any conversion event, such as Search, View Content, Add to Cart and Purchase.
- Conversion Rate: Number of Actions / Number of Website Clicks x 100


Definition of the Necesarry Modules


```python
import pandas as pd
from scipy.stats import shapiro
import scipy.stats as stats
import numpy as np
import seaborn as sns
from scipy.stats import shapiro
import matplotlib.pyplot as plt
```

Importing the Datasets: Two sets are stored in different sheets of excel

```python
A = pd.read_excel("../input/ab-testing/ab_testing_data kopyas 2.xlsx" , sheet_name= "Control Group")
B = pd.read_excel("../input/ab-testing/ab_testing_data kopyas 2.xlsx" , sheet_name= "Test Group")
```

Selection of the Metric Variable: It is specified that "Purchase" column must be considered

```python
A = A[["Purchase"]]
B = B[["Purchase"]]
```

Combining the Data Sets

```python
A_B = pd.concat([A, B], axis=1)
A_B.columns = ["A", "B"]
```

Generating the Group Index of A

```python
GRUP_A = np.arange(len(A))
GRUP_A = pd.DataFrame(GRUP_A)
GRUP_A[:] = "A"
A = pd.concat([A, GRUP_A], axis=1)
```

Generating the Group Index of B

```python
GRUP_B = np.arange(len(B))
GRUP_B = pd.DataFrame(GRUP_B)
GRUP_B[:] = "B"
B = pd.concat([B, GRUP_B], axis=1)
```

Final Version of the Data Set

```python
AB = pd.concat([A, B])
AB.columns = ["Purchase", "Group"]
```

Crosstable of Summary Statistics by Groups

```python
AB.groupby("Group").aggregate(["mean","std","min",max])
```

Testing the Assumptions of Independent Samples t-Test
1.Normality
2.Homogenty

In this part of the study it is necesarry to decide which way of testing the significany of difference will be choosen.
Shaphiro Wilks test is a way to test if the serie has normal distrubation or not.
Also a Box Plot might be helpfull to observe if the serie has outliers or not.

Box Plot


```python
sns.boxplot(x="Group", y="Purchase", data=AB);
plt.show()
```

Shaphiro Wilks Test Hypothesises

H0: The sample data are not significantly different than a normal
population.

Ha: The sample data are significantly different than a normal
population.

```python
shapiro(A_B.A)
```

```python
shapiro(A_B.B)
```

As can be seen above p > .05 which indicates that H0 can NOT be rejected. Accordingly distrubations of each serie is likely to normal distrubation. We are able to run t Test.

Levene Test of Homogenity

H0: The variances are equal across all samples.

Ha: The variances are not equal for at least one pair.

```python
stats.levene(A_B.A, A_B.B)
```

As can ve seen above p > .05 which indicates that H0 can NOT be rejected. Accordingly variances are equal and we need to run the t Test accordingly.


Time to Run Independent Samples of t-Test

#1. How would you define the hypothesis of this A/B test?


- H0: There is NO statistically significant difference between A and B groups in Purchase values.

- Ha: There iş a statistically significant difference between A and B groups in Purchase values.
    
    


```python
tvalue, pvalue = stats.ttest_ind(A_B["A"], A_B["B"], equal_var=True)
print('tvalue = %.4f, pvalue = %.4f' % (tvalue, pvalue))
```

As can be seen above p > .05 which indicates that H0 can NOT be rejected. Accordingly there is no difference between groups in the case of Purchase values.

#2. Can we conclude statistically significant results?
- Two independent samples test results greater than 0.05 using test and control group data set H0 hypothesis cannot be rejected because of its outcome. In other words, the purchase made with the maximum bidding and average bidding methods It was observed that there was no significant difference between the averages of the intake.
